```python
import pandas as pd
from matplotlib import pyplot as plt
from numpy import sqrt
from scipy.stats import norm
import numpy as np

#Question
#A common perception about COVID-19 is that Warm Climate is more resistant to the corona outbreak and we need to
#verify this using Hypothesis Testing. So what will our null and alternate hypothesis be?

hello=pd.read_csv('E:\Python\Corona_Updated.csv')

#Ho=warm climate(temperature) not effect to corona outbreak
#Ha=warm climate(temperature) effect the corona outbreak

hello.head()


data1=hello[hello['Temp_Cat']==1]['Confirmed']
data2=hello[hello['Temp_Cat']==0]['Confirmed']


m1=np.mean(d1)
m2=np.mean(d2)
n1=len(d1)
n2=len(d2)
sigma1=np.std(d1)
sigma2=np.std(d2)


def corona(m1,m2,n1,n2,sigma1,sigma2):
    z= ((m1-m2)/sqrt((sigma1**2/n1)+(sigma2**2/n2)))
    pval = 2*(1 - norm.cdf(abs(z)))
    return pval



pvalue=corona(m1,m2,n1,n2,sigma1,sigma2)
alpha=0.05

if pvalue>0.05:
    print('Accept null hypothesis')
    print('Temperature does not effect the corona')
else:
    print('Reject null hypothesis')
    print('Temperature does effect the corona')
```
